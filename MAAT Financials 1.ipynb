## Notes on Cost Scalability and Marginal Cost Structure

MAAT’s cost architecture is explicitly designed to preserve capital and scale efficiently under regulatory and infrastructural constraints. The following five principles govern the current cost structure:

### 1. Low Marginal Cost per User

- Identity verification and transactional costs scale directly with user activity (e.g., €1–2 per KYC check), ensuring that costs are incurred only upon engagement.
- Custody infrastructure, such as BitGo or Fireblocks, follows transaction-based pricing or fixed monthly thresholds, aligning operational expenses with actual system usage.

### 2. Infrastructure Cost Minimisation

- Core infrastructure tools—including The Graph, IPFS/Filecoin, Cloudflare R2, and AWS micro-instances—offer generous free tiers or ultra-low per-unit costs suitable for MVP deployment.
- Data indexing and storage remain negligible in cost (e.g., $0.00004 per query or €0.015/GB) until operational scale is achieved, allowing capital expenditure to remain deferred.

### 3. Legal and Compliance Containment

- Legal and compliance costs are front-loaded but non-recurring. Key regulatory elements such as GDPR documentation and MiCA-readiness are budgeted as amortised one-time costs.
- This structure enables fixed budgeting for foundational compliance without prematurely locking into burdensome licensing regimes (e.g., full BaFin crypto licensing).

### 4. Engineer-Led Efficiency

- Technical delivery is driven by two part-time engineers (backend and blockchain). This configuration supports the MVP’s requirements without engaging in full-stack or front-end development at this stage.
- No permanent operational staff or DevOps resources are retained, further reducing structural cost obligations during the pilot phase.

### 5. Pass-Through Design Philosophy

- Cost components associated with user activity (e.g., identity verification, custody transfers) are eligible for full or partial pass-through pricing in future institutional contracts.
- The infrastructure is designed for white-label deployment, enabling recurring revenue opportunities that naturally absorb or offset variable platform costs.

### Conclusion

MAAT’s budget demonstrates a high fixed-to-variable cost leverage ratio, strong alignment between expenditure and usage, and minimal capital lock-in. This structure ensures operational flexibility during pilot and pre-revenue phases and enables credible runway management for early-stage investment.

- Avg. B2B client pays: €500/month (compliance + identity + reporting package)
- Monthly burn: €9,950
- Gross margin: ~80% on subscription-based model

Estimated Break-Even Point:
- Minimum 25 enterprise users required to fully cover burn
- Break-even revenue = €10,000/month
- Break-even timeline: Month 10–12 if onboarding accelerates post-pilot

Note:
This does not assume upsell, white-labelling revenue, or usage-based pass-throughs, which could reduce break-even further.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assumptions
months = list(range(1, 25))
price_per_client = 500
gross_margin = 0.80
monthly_burn = 9950
initial_clients = 1
growth_rate = 0.30  # 30% growth per month

# Simulate client growth and financials
clients = [initial_clients]
revenues = [initial_clients * price_per_client * gross_margin]
burns = [monthly_burn]
profits = [revenues[0] - burns[0]]
cumulative_profit = [profits[0]]

for i in range(1, len(months)):
    new_clients = clients[-1] * (1 + growth_rate)
    new_revenue = new_clients * price_per_client * gross_margin
    new_profit = new_revenue - monthly_burn
    new_cumulative = cumulative_profit[-1] + new_profit

    clients.append(new_clients)
    revenues.append(new_revenue)
    burns.append(monthly_burn)
    profits.append(new_profit)
    cumulative_profit.append(new_cumulative)

# Find break-even month
breakeven_month = next((i + 1 for i, cp in enumerate(cumulative_profit) if cp >= 0), None)

# Create DataFrame
df = pd.DataFrame({
    "Month": months,
    "Clients": [int(c) for c in clients],
    "Monthly Revenue (€)": revenues,
    "Monthly Burn (€)": burns,
    "Monthly Profit (€)": profits,
    "Cumulative Profit (€)": cumulative_profit
})

# Plot cumulative profit
plt.figure(figsize=(10, 6))
plt.plot(months, cumulative_profit, marker='o', label='Cumulative Profit (€)', color='darkblue')
plt.axhline(0, color='red', linestyle='--', label='Break-Even Threshold')
if breakeven_month:
    plt.axvline(x=breakeven_month, color='gray', linestyle='--', label=f'Break-Even: Month {breakeven_month}')
plt.title("Break-Even Timeline – MAAT MVP", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Cumulative Profit (€)")
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# === Constants ===
months = np.arange(1, 30)
monthly_burn = 9950
gross_margin = 0.80
initial_clients = 1

# === Define Scenarios ===
scenarios = [
    {"label": "30% Growth @ €500", "growth": 0.30, "price": 500, "color": "navy"},
    {"label": "20% Growth @ €500", "growth": 0.20, "price": 500, "color": "darkgreen"},
    {"label": "30% Growth @ €750", "growth": 0.30, "price": 750, "color": "darkred"}
]

# === Break-even summary list ===
breakeven_summary = []

# === Loop through scenarios ===
for scenario in scenarios:
    label = scenario["label"]
    color = scenario["color"]

    # Simulate clients and MRR
    clients = [initial_clients]
    mrr = [initial_clients * scenario["price"] * gross_margin]

    for _ in range(1, len(months)):
        new_clients = clients[-1] * (1 + scenario["growth"])
        clients.append(new_clients)
        mrr.append(new_clients * scenario["price"] * gross_margin)

    # Find break-even
    for i, revenue in enumerate(mrr):
        if revenue >= monthly_burn:
            breakeven_month = months[i]
            breakeven_mrr = revenue
            break
    else:
        breakeven_month = None
        breakeven_mrr = None

    # Store break-even info
    breakeven_summary.append({
        "Scenario": label,
        "Break-even Month": breakeven_month if breakeven_month else "Not reached",
        "MRR at Break-even (€)": round(breakeven_mrr, 2) if breakeven_mrr else "N/A"
    })

    # Plot individual chart
    plt.figure(figsize=(8, 5))
    plt.plot(months, mrr, marker='o', color=color, linewidth=2)
    plt.axhline(y=monthly_burn, color='gray', linestyle='--', linewidth=1)
    plt.title(label, fontsize=14, fontweight='bold')
    plt.xlabel("Month")
    plt.ylabel("Monthly Recurring Revenue (€)")
    plt.grid(True, linestyle='--', linewidth=0.5)

    # Annotate break-even
    if breakeven_month:
        plt.scatter(breakeven_month, breakeven_mrr, color=color, edgecolor='black', zorder=5)
        plt.annotate(f"Break-even\nMonth {breakeven_month}",
                     (breakeven_month, breakeven_mrr + 800),
                     ha='center', fontsize=9, color=color)

    plt.tight_layout()
    plt.show()

    # Save CSV for this scenario
    df = pd.DataFrame({
        "Month": months,
        "Clients": clients,
        "MRR (€)": mrr
    })
    safe_label = label.replace(" ", "_").replace("@", "at").replace("€", "EUR")
    df.to_csv(f"MRR_{safe_label}.csv", index=False)

# === Save Break-even Summary ===
df_summary = pd.DataFrame(breakeven_summary)
df_summary.to_csv("MRR_Breakeven_Summary.csv", index=False)
```

## Financial Projection: Cost Structure, Revenue Growth, and Cash Flow Dynamics

This section presents a month-by-month simulation of MAAT’s projected financial position during its early growth phase. The analysis reflects conservative operational assumptions and applies institutional standards of solvency monitoring. The objective is to evaluate the sustainability of the MVP phase and identify the inflection point at which monthly revenue offsets operating costs.

### Assumptions

| Parameter                          | Value          |
|-----------------------------------|----------------|
| Initial Monthly Operating Cost    | €9,950         |
| Opening Cash Reserves             | €100,000       |
| Starting Number of Clients        | 1              |
| Monthly Client Growth Rate        | 30% compounded |
| Monthly Revenue per Client        | €500           |
| Gross Margin                      | 80%            |
| Projection Horizon                | 18 months      |

---

### Methodology

The simulation proceeds under the following structure:

1. **Client Growth Dynamics**  
   $$
   \text{Clients}_t = \text{Clients}_{t-1} \times (1 + 0.30)
   $$

2. **Revenue Estimation**  
   $$
   \text{Revenue}_t = \text{Clients}_t \times 500 \times 0.80
   $$

3. **Monthly Net Cash Flow**  
   $$
   \text{Net Cash Flow}_t = \text{Revenue}_t - \text{Burn}
   $$

4. **Cumulative Cash Position**  
   $$
   \text{Cash}_t = \text{Cash}_{t-1} + \text{Net Cash Flow}_t
   $$

---

### Observations

- Under the specified conditions, MAAT maintains a positive cash position for approximately thirteen months.
- Break-even is achieved in Month 14, at which point cumulative net cash flow surpasses cumulative burn.
- Monthly revenue is non-linear, driven by geometric client growth; this offsets fixed costs without assuming variable overhead escalation.
- The model presumes no further equity injection post-raise and is designed to reflect capital discipline in a pre-revenue context.

---

### Conclusion

This simulation validates the financial structure of MAAT’s MVP phase.
